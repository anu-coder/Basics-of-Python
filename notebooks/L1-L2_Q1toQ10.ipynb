```python
def generate_dict(n: int):
    out_dict = {}
    for i in range(1,n+1):
        out_dict[i] = i*i
    return out_dict
```

```python
print(generate_dict(8))
```

```python
def list_sort_arrange(sent):
    words = sent.split(' ')
    check = []
    for word in words:
        if word not in check:
            check.append(word.lower())
    check.sort()
    return check
```

```python
list_sort_arrange('hello world and practice makes perfect and hello world again')
```
